```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario_marcos = np.array([100, 200, 400, 300])
salario_gileno = np.array([50, 300, 500, 450])
```

```python
plt.plot(salario_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salario_gileno, c='Red', ls='--', marker='s', ms=8, label='Salário Gileno')
plt.show
```

```python
#Com Legenda
plt.plot(salario_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salario_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.legend()
plt.show
```

```python
#Com Legenda
plt.plot(salario_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salario_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.legend(loc='upper right')
plt.show
```

```python
salario_pedro = [200, 150, 300, 700]
plt.plot(salario_marcos, c='Black', ls='--', marker='s', ms=8, label='Salário Marcos')
plt.plot(salario_gileno, c='Red', ls='--', marker='^', ms=8, label='Salário Gileno')
plt.plot(salario_pedro, c='Green', ls='--', marker='o', ms=8, label='Salário Pedro')
plt.legend(loc='upper left')
plt.show
```
